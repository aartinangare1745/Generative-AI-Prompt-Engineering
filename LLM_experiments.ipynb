```python
from groq import Groq
client=Groq()

response=client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    max_tokens=300,
    messages=[
        {"role":"system","content":"you are friendly assistant helping someone to learn code"},
        {"role":"user","content":"in 3 sentence, explain what is API,like I am new to programming"}
    ]

)

print(response.choices[0].message.content)
```

```python
# PROJECT 1: A Real Chatbot Loop, back and forth conversation
```

```python
from groq import Groq
client=Groq()
conversation=[{"role":"system","content":"you are a friendly assistant"}]
print("chat with the model. type 'quit' to stop.\n")
```

```python
while True:
  user_input=input("You: ")

  if user_input.lower()=="quit":
    print("bye! ")
    break
  conversation.append({"role":"user","content":user_input})

  #send entire history every single time.costs more tokens for long conversation
  response=client.chat.completions.create(
      model="llama-3.3-70b-versatile",
      max_tokens=300,
      messages=conversation
  )
  reply=response.choices[0].message.content
  print(f"model: {reply}\n")

  #add models reply to history too.
  conversation.append({"role":"assistant","content":reply})
```
